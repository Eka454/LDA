## Данные для расчета инфомации (матрицы phi и theta, набор тем)

```python
phi_100 = lda.phi_ # вероятность термов W в каждой теме t
phi_T = phi_100.T.to_dict('list')
change_keys = [key[1] for key in list(phi_T.keys())]
phi = dict(zip(change_keys, list(phi_T.values()))) # для расчетов
```

```python
theta_100 = pd.read_pickle("theta_100.pkl") # вероятность тем t в каждом документе D
theta = np.array(theta_100[range(0, theta_100.shape[1])].T) # для расчетов
```

## Данные по каждому документу.  
* Выделить топ 2 темы для каждого документа  
* Выделить топ 3 слова для документа в каждой теме

```python
ann = pd.read_pickle('annotations.pkl')  
data = pd.read_pickle('tokens.pkl')  
data.drop(data[data['tokens'].isna()].index, inplace=True)
data.index = np.arange(len(data))
```

```python
def top_n(list_sort, n): # возвращает индексы top n значений
    try:
        return np.argpartition(
            list_sort, np.arange(-n, 0), axis=-1
            )[..., :-n-1:-1]
    except ValueError:
        return top_n(list_sort, n-1)
```

```python
tems = top_n(theta, 2)
d_t_3w = {}
for d in range(data.shape[0]):
    d_t_3w[d] = []
    for tem in tems[d]:
        words = np.unique(np.array(data.tokens[d].split()))
        word_in_doc = []
        for word in words:
             word_in_doc.append(phi[word][tem])
        top_3 = top_n(word_in_doc, 3)
        d_t_3w[d].append([tem, words[top_3]])  # номер документа => [номер темы, топ 3 слова в данном документе для данной темы]
```

## Проверить результаты для отдельных документов 

```python
n = 41098 # номер документа
info = d_t_3w[n] # info о документе n
for i in info:
    t = i[0]
    print('Topic #{0}: {1}'.format(t, top_tokens[t])) # вывод info
number_in_annotations = data['Unnamed: 0'][n]
text = ann.annotation[number_in_annotations] # содержание текста
text_in_t = data.tokens[n] 
print('\n', text)
print('\n', text_in_t)
```
